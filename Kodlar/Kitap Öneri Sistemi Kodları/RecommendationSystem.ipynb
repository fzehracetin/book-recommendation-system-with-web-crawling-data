```python
users = pd.read_csv('user_data.csv', sep=',', error_bad_lines=False, encoding="utf-8")
users.columns = ['User Id', 'Yas', 'Cinsiyet']

books = pd.read_csv('book_file.csv', sep=',', error_bad_lines=False, encoding="utf-8")
books.columns = ['Adı', 'Yazar', 'Alt başlık', 'Baskı tarihi', "Sayfa sayısı", "Format", "ISBN", "Kitabın türü", "Orijinal adı",
                  "Çeviri", "Dil", "Ülke", "Yayınevi", "Baskılar", "Yaş", "Cinsiyet", "Puan", "Oy Sayısı"]

ratings = pd.read_csv('rating_data.csv', sep=',', error_bad_lines=False, encoding="utf-8")
ratings.columns = ['User Id', 'ISBN', 'Puan']
```

Rating tablosu üzerinden en çok oy alan ilk 5 kitabın ISBN'lerinin gösterilmesi.

```python
rating_count = pd.DataFrame(ratings.groupby('ISBN')['Puan'].count())
rating_count.sort_values('Puan', ascending=False).head()
```

En çok oy alan ilk 5 kitabın ISBN'leri kullanılarak kitap verisi üzerinden kitapların bilgilerinin yazdırılması.

```python
most_rated_books = pd.DataFrame(['9789755105857', '9789754731927', '9786053320937', '9789751403643', '9786053111795'], index=np.arange(5), columns = ['ISBN'])
most_rated_books_summary = pd.merge(most_rated_books, books, on='ISBN')
most_rated_books_summary
```

En çok oy verilen kitapların ortalama puanlarının bulunması.

```python
average_rating = pd.DataFrame(ratings.groupby('ISBN')['Puan'].mean())
average_rating['ratingCount'] = pd.DataFrame(ratings.groupby('ISBN')['Puan'].count())
average_rating.sort_values('ratingCount', ascending=False).head()
```

```python
ratings.set_index(['User Id', 'ISBN', 'Puan'], append=True)
```

Ratings tablosundaki değerlerin 1000x999 boyutlu matrise aktarılması. 

Boyut: Kullanıcı sayısı x Kitap sayısı  

```python
ratings_pivot = ratings.pivot(index='User Id', columns='ISBN').Puan
userID = ratings_pivot.index
ISBN = ratings_pivot.columns
print(ratings_pivot.shape)
ratings_pivot.head()
```

Bu fonksiyon parametre olarak aktarılan ISBN değerine en benzer kitapların benzerlik değerini içeren tabloyu döndürür.

```python
def isbn_finder(isbn):
    bones_ratings = ratings_pivot[isbn]
    similar_to_bones = ratings_pivot.corrwith(bones_ratings)
    corr_bones = pd.DataFrame(similar_to_bones, columns=['pearsonR'])
    corr_bones.dropna(inplace=True)
    corr_summary = corr_bones.join(average_rating['ratingCount'])
    corr_summary[corr_summary['ratingCount']>=50].sort_values('pearsonR', ascending=False)
    return corr_summary
```

Bu fonksiyon parametre olarak ISBN değeri alır, dönüş değeri parametre olarak aktarılan kitaba en benzer 5 kitaptır.

```python
def corr_finder(isbn):
    corr_sum = isbn_finder(isbn)
    corr_sum = corr_sum[corr_sum['ratingCount']>=50].sort_values('pearsonR', ascending=False)

    books_corr_to_bones = pd.DataFrame([corr_sum.index[1], corr_sum.index[2], corr_sum.index[3], 
                                        corr_sum.index[4], corr_sum.index[5]], 
                                  index=np.arange(5), columns=['ISBN'])
    corr_books = pd.merge(books_corr_to_bones, books, on='ISBN')
    return corr_sum, corr_books
```

Fonksiyonun çağırılması ve kitapların bastırılması.

```python
corr_sum, corr_books = corr_finder('9789944880985')
corr_sum.head(5)
```
