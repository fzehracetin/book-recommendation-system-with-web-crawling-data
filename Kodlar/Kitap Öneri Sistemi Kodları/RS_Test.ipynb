```python
with io.open('test_data.csv', mode='w', encoding="utf-8") as test_file:
    fieldnames = ['User Id', 'ISBN', 'Puan']
    test_writer = csv.DictWriter(test_file, fieldnames=fieldnames)
    test_writer.writeheader()

    test_dict = {
        "User Id": "null",
        "ISBN": "0",
        "Puan": "null"
    }
    
    for i in range(200):      
        for j in range(50, 100):
            test_dict['ISBN'] = ratings['ISBN'][(index[i]-1)*100 + j]
            test_dict['User Id'] = ratings['User Id'][(index[i]-1)*100 + j]
            test_dict['Puan'] = ratings['Puan'][(index[i]-1)*100 + j]
            ratings = ratings.drop([(index[i]-1)*100 + j],axis = 0)
            test_writer.writerow(test_dict)
```

En çok oy verilen kitapların ortalama puanlarının bulunması.

```python
average_rating = pd.DataFrame(ratings.groupby('ISBN')['Puan'].mean())
average_rating['ratingCount'] = pd.DataFrame(ratings.groupby('ISBN')['Puan'].count())
average_rating.sort_values('ratingCount', ascending=False).head()
```

```python
ratings.set_index(['User Id', 'ISBN', 'Puan'], append=True)
```

Ratings tablosundaki değerlerin 1000x999 boyutlu matrise aktarılması. 

Boyut: Kullanıcı sayısı x Kitap sayısı  

```python
ratings_pivot = ratings.pivot(index='User Id', columns='ISBN').Puan
userID = ratings_pivot.index
ISBN = ratings_pivot.columns
print(ratings_pivot.shape)
ratings_pivot.head()
```

Bu fonksiyon parametre olarak aktarılan ISBN değerine en benzer kitapların benzerlik değerini içeren tabloyu döndürür.

```python
def corr_finder(isbn):
    bones_ratings = ratings_pivot[isbn]
    similar_to_bones = ratings_pivot.corrwith(bones_ratings)
    corr_bones = pd.DataFrame(similar_to_bones, columns=['pearsonR'])
    corr_bones.dropna(inplace=True)
    corr_summary = corr_bones.join(average_rating['ratingCount'])
    corr_summary[corr_summary['ratingCount']>=50].sort_values('pearsonR', ascending=False)
    return corr_summary
```

Bu fonksiyon tavsiye edilen kitaplar ile test kümesindeki kitapları kıyaslayarak çakışma durumunda ortalama değeri ve çakışma sayısını gönderir.

```python
def compare(recommendations, i):
    count = 0
    avg = 0
    for tavsiye in recommendations:
        # print(tavsiye)
        flag = 0
        j = 0
        while (j < 50 and not flag):
            # print(j)
            if(test['ISBN'][i*50 + j] == tavsiye):
                print(test['Puan'][i*50 + j])
                flag = 1
                count += 1
                avg += test['Puan'][i*50 + j]
            j += 1
    return (avg, count)
```

Bu kod parçası test kümesindeki her bir kullanıcı için tavsiyeler üretir. Üretilen tavsiyeler ile test kümesinde çakışmalar olması durumunda ortalama değeri hesaplayarak tavsiye sisteminin başarısını ölçer.

```python
avg = 0
count = 0
for i in range(200):
    recommendations = []
    avg_corr = 0
    for j in range(50):
        if (ratings['Puan'][(index[i]-1)*100 + j] >= 7):
            corr_sum = corr_finder(ratings['ISBN'][(index[i]-1)*100 + j])
            corr_sum = corr_sum[corr_sum['ratingCount']>=50].sort_values('pearsonR', ascending=False)
            for k in range(1,6):
                if corr_sum.index[k] not in recommendations: 
                    recommendations.append(corr_sum.index[k])
    avg_tmp, count_tmp = compare(recommendations, i)
    avg += avg_tmp           
    count += count_tmp 
print(avg)
print(count)
print(avg/count) 
```
